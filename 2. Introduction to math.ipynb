## 1. Linear Function

The shape of it is just a straight line

### Properties 

The useful form of the linear equation would be slope and y-interception form: 

$$ f(x) = mx + c $$

Where m is a slope of the graph and c is an y-intercept of the graph.

### What is the slope?

Slope is the value that tells us how steep the graph is or the more useful definition: how much y changes when x changes in certain values.

#### How can we create slopes?

By using this equation: $ m = \frac{\Delta y}{\Delta x} $ (We call this triangle delta)

Delta X is the change in X values and delta Y is the change in Y values.

We can say that the line that has the slope of 2 means that when x is increased by 1, y is going to increase by 2, which creates the property of steepness.

Furthermore, if the slope is -2, this means Y is going to decrese by 2 if X is increased by 1.

#### For Example

If we got 4 sets of points : 

| No. of Dot | X value | Y value |
| ---- |:------:|------:|
| 1. | 5 | 5 |
| 2. | 15 | 15 |
| 3. | 15 | 20 |

And we want to find the slope between point 1 and 2 and between point 1 and 4.

The slope between point 1 and 2 would be $ \frac{15-5}{15-5} = 1 $ and slope between point 1 and 3 would be $ \frac{20-5}{15-5} = 1.5 $. 

```python
# List all of the points

point1 = [5,5]
point2 = [15,15]
point3 = [15,20]

# Define the line plot between 2 points
def plot_points(p1, p2):
    plt.plot([p1[0], p2[0]], [p1[1], p2[1]])

# Plot 2 lines 
plot_points(point1, point2)
plot_points(point1, point3)

# Adding Grid
plt.grid()

plt.axis([0, 20, 0, 25])
plt.show()
```

As you can see, the green line (point 1 and 3) is steeper than the blue line (point 1 and 2). Slope can be very useful in the future.

## 2. Polynomial

Polynomial consists of many power terms add together. For example, $ 3x^2 + 2, 4x^3 + 2x^2 + x + 1 $.
The degree can't be anything that is less than or equal to 1.

### Before introduce the general formula

Lets talks about the sigma symbol.

Sigma symbol($\sum $) is associated with sum. 

#### For example

$\sum_{n=1}^{5} n = 1+2+3+4+5 $

$\sum_{n=1}^{5} 2n = 2*1+2*2+2*3+2*4+2*5 $

The number at the top tells us the limit and the number at the bottom tells us the initial value also the symbol "n" is the placeholder that we can add with numbers between the top and the start and we can do anything with it.

### General formula

The degree of the polynomial is the highest power in that polynomial. 

The general formula for n degree would be: 

$$ \sum_{i=0}^{n} c_i * x^i $$

where $ c_i $ is just a constant (any real number).

#### For example

$$ \sum_{i=0}^{3} c_i * x^i = c_0+ c_1x + c_2x^2  + c_3x ^ 3 $$

and $ c_0, c_1, c_2, c_3 $ can be any number you want (it can be zero too. This will cause that term to be omitted because anything that is multiplied by 0 will be zero).

### The shape of polynomial

The shape is associated with the degree of the polynomial. The even degree of the polynomial will look similar, same as the odd degree.

### Even degree.

```python
# make the curve looks real not just lines connecting with points
import numpy as np

# Input of the graph
input_graph = np.arange(-5, 5, 0.25)

# Using Lambda
function1 = lambda x : x**2
function2 = lambda x : x**4
function3 = lambda x : x**6

# Applying the function to the input to get output
output_graph1 = function1(input_graph)
output_graph2 = function2(input_graph)
output_graph3 = function3(input_graph)

# difference graph in one plot
plt.plot(input_graph, output_graph1)
plt.plot(input_graph, output_graph2)
plt.plot(input_graph, output_graph3)

plt.xlabel('Inputs')
plt.ylabel('Outputs')

# Adding Grid
plt.grid()

plt.axis([-10, 10, -5, 10])
plt.show()
```

### Odd degree.

```python
# Input of the graph
input_graph = np.arange(-5, 5, 0.25)

# Using Lambda
function1 = lambda x : x**3
function2 = lambda x : x**5
function3 = lambda x : x**7

# Applying the function to the input to get output
output_graph1 = function1(input_graph)
output_graph2 = function2(input_graph)
output_graph3 = function3(input_graph)

# difference graph in one plot
plt.plot(input_graph, output_graph1)
plt.plot(input_graph, output_graph2)
plt.plot(input_graph, output_graph3)

plt.xlabel('Inputs')
plt.ylabel('Outputs')

# Adding Grid
plt.grid()

plt.axis([-5, 5, -5, 5])
plt.show()
```

## Derivative.

First of all, I am not going through all the details. If you want to learn more about this, I suggests [Khan Accademy](https://www.khanacademy.org).

Derivative is all about the slope of the graph at one point. The straight line that touches one point on a graph and has the same slope is called the tangent line.

![Tangent line from paul's online math note](http://tutorial.math.lamar.edu/Classes/CalcI/Tangents_Rates_files/image001.gif)

### Example

If we want to find the slope of a graph $ f(x) = 2x^2 $ at $x = 1$, then we have to find the function that correspondedto the slope of the graph $ f(x) = 2x^2 $. We use $\frac{d}{dx} $ to get the derivative that respect to X (We are interested in the change of output when x changes because there may be more that one variable that affect the function)

### How we calculate?

1. Power rule : $\frac{d}{dx} x^n = nx^{n-1}$. For example, $\frac{d}{dx} x^2 = 2 * x^{2-1} = 2x$ 
2. Constant(any number) : $\frac{d}{dx} C = 0 $
3. Sum : $\frac{d}{dx} [f(x) + g(x)] = \frac{d}{dx} f(x) + \frac{d}{dx} g(x) $. For example $\frac{d}{dx} x^2 + x^3 = 2x + 3x^2$ 
4. Minus : $\frac{d}{dx} [f(x) - g(x)] = \frac{d}{dx} f(x) - \frac{d}{dx} g(x) $. For example $\frac{d}{dx} 3x^2 - x^3 = 6x - 3x^2$ 
5. Product rule : $\frac{d}{dx} [f(x) * g(x)] = \frac{d}{dx} f(x) * g(x) + \frac{d}{dx} g(x) * f(x) $. For example $\frac{d}{dx} 3x^2*x^3 = 6x*x^3 + 3x^2*3x^2 = 6x^4 + 9x^4$ 
6. Chain rule : $\frac{d}{dx} [f(g(x))] = \frac{d}{dg(x)} f(g(x)) \frac{d}{dx} g(x) $. For example $\frac{d}{dx}(2x + 3)^6 = 6(2x+3)^5 * 2 = 12(2x+3)^5 $. This rule is very useful because you don't have to expand the polynomial, which will take a lot of time.

### Why it is useful?

Derivative helps us to find the critical point where it can be the minimum point or maximum point in the graph or just a flat area. All of these points are called critical points. 

### What dothey have in common? 

The answer is the slope at that point is zero. Can you picture it thinking about the tangent line? If you can't, take a look at the polynomial graph. For the even degree polynomials, the lowest point is at zero. And, for the odd degree polynomial, at zero, it is not the maximum point nor minimum point but it is the flat area.

### How do we know that that point is maximum point or minimum point?

We use the second derivative (differentiate the derivative again). It will tell how the first derivative behaves. If the second derivative is positive, this means that the slope of first derivative is positive, which means that the first derivative is incresed and vice versa. 

```python
# For Minimum point
input_graph = np.arange(-5, 5, 0.25)

# Using Lambda
function = lambda x : x**2

# Calulate the tagent line by hand
tangent_line_1 = lambda x : -2*x - 1
tangent_line_2 = lambda x : -4*x - 4
tangent_line_3 = lambda x : 2*x-1

# Plot the graph
plt.plot(input_graph, function(input_graph))
plt.plot(input_graph, tangent_line_1(input_graph))
plt.plot(input_graph, tangent_line_2(input_graph))
plt.plot(input_graph, tangent_line_3(input_graph))

plt.xlabel('Inputs')
plt.ylabel('Outputs')

# Adding Grid
plt.grid()

plt.axis([-5, 5, -5, 5])
plt.show()
```

```python
# For Maximum point

# Using Lambda
function = lambda x : -x**2

# Calulate the tagent line by hand
tangent_line_1 = lambda x : 2*x + 1
tangent_line_2 = lambda x : 4*x + 4
tangent_line_3 = lambda x : -2*x + 1

# Plot the graph
plt.plot(input_graph, function(input_graph))
plt.plot(input_graph, tangent_line_1(input_graph))
plt.plot(input_graph, tangent_line_2(input_graph))
plt.plot(input_graph, tangent_line_3(input_graph))

plt.xlabel('Inputs')
plt.ylabel('Outputs')

# Adding Grid
plt.grid()

plt.axis([-5, 5, -5, 5])
plt.show()
```

For the minimum point, the slope is increasing when x increases, so for all minimum points the second derivative of the graph will be positive. And for the maximum point, it is the opposite.

### Why they are useful?

I will give you the example of the simple optimization problem. The problem is that we are going to find the highest profit, where profit is repesented with this function $ P(x) = x^3-5x^2+2x+10 $ without plotting the graph. 

First we have to find the critical point. 

We first use the sum rule : $\frac{d}{dx} P(x) =  \frac{d}{dx} x^3 - \frac{d}{dx} 5x^2 + \frac{d}{dx} 2x + \frac{d}{dx} 10 = 3x^2 - 10x + 2$

In order to find the critical point, we have to get the derivative to be zero:

$3x^2 - 10x + 2 = 0$, so $x = \frac{\sqrt{19} + 15}{3} \approx 0.214, \frac{- \sqrt{19} + 15}{3} \approx 3.12$ (I won't go through the calculation, this link will tell you how I find the solution[this](https://www.khanacademy.org/math/algebra/quadratics/solving-quadratics-using-the-quadratic-formula/v/using-the-quadratic-formula))

In order to find the maximum, we have to differentiate again:
$\frac{d}{dx} 3x^2 - 10x + 2 = \frac{d}{dx} 3x^2 - \frac{d}{dx} 10x + \frac{d}{dx} 2 = 6x - 10$

Then we plug the x-coordinates of critical points to the second derivative to get the behavior of the first derivative.

$6(0.214) - 10 = -8.716$ and $6(3.12) - 10 = 8.72$

So yjr point at $x = 0.214$ is the maximum point, while the point at $x = 3.12$ is the minimum point. The maximum profit would be $ (0.214)^3-5(0.214)^2+2(0.214)+10 = 10.29 $

### Lets check the answer

```python
# Check the answer
input_graph = np.arange(-5, 5, 0.25)

# Using Lambda
profit = lambda x : x**3 - 5*(x**2) + 2*x + 10

# Plot the graph
plt.plot(input_graph, profit(input_graph))

# Plot the max point too
plt.plot([0.214],[10.29], 'ro')

plt.xlabel('Inputs')
plt.ylabel('Profit')

# Adding Grid
plt.grid()

plt.axis([-5, 5, 0, 15])
plt.show()
```
